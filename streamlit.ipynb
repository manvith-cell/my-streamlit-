### 2. Create the Streamlit App File
Next, we'll create a Python file named `streamlit_app.py` containing a simple Streamlit application. We'll use the `%%writefile` magic command for this.

```python
%%writefile streamlit_app.py

import streamlit as st

st.title("Hello from Colab Streamlit!")
st.write("This is a simple Streamlit app running in Google Colab.")

name = st.text_input("What's your name?")
if name:
    st.write(f"Hello, {name}!")
```

### 3. Create `requirements.txt` (Optional)
If your Streamlit app has additional dependencies, you would list them in a `requirements.txt` file. For this simple example, we only need Streamlit, which we already installed. We'll create an empty one for demonstration.

```python
%%writefile requirements.txt
streamlit
```

### 4. Run the Streamlit App
Finally, we'll run the Streamlit app. Colab often requires `npx localtunnel` or `cloudflared` to create a public URL for your local Streamlit server. This command will install `localtunnel` and then run your Streamlit app, providing a public URL you can click to access it.
